## 1. Introduction <a id='1-introduction'></a>

The **Markowitz Portfolio Optimization** framework, also known as **Modern Portfolio Theory (MPT)**, is a mathematical framework for constructing a portfolio of assets such that the expected return is maximized for a given level of risk. Introduced by Harry Markowitz in his seminal 1952 paper "Portfolio Selection," this theory revolutionized the field of finance by formalizing the intuitive concept of diversification.

The central insight is that an asset's risk and return should not be assessed in isolation, but rather by how it contributes to a portfolio's overall risk and return. This led to the concept of the **efficient frontier**â€”the set of portfolios that offer the highest expected return for each level of risk.

---

## 2. Historical Background <a id='2-historical-background'></a>

Before Markowitz, investment decisions were primarily based on individual security analysis. Investors focused on finding "good" stocks without a systematic approach to portfolio construction.

**Key milestones:**

- **1952:** Harry Markowitz publishes "Portfolio Selection" in the *Journal of Finance*, introducing mean-variance optimization.
- **1959:** Markowitz expands his work in the book *Portfolio Selection: Efficient Diversification of Investments*.
- **1964-1966:** William Sharpe, John Lintner, and Jan Mossin independently develop the Capital Asset Pricing Model (CAPM), building on Markowitz's work.
- **1990:** Harry Markowitz receives the Nobel Prize in Economics (shared with Merton Miller and William Sharpe) for his pioneering work in the theory of financial economics.

## 3. Mathematical Framework <a id='3-mathematical-framework'></a>

### 3.1 Basic Definitions

Consider a portfolio with $n$ risky assets. Let:

- $\mathbf{w} = (w_1, w_2, \ldots, w_n)^T$ be the vector of portfolio weights
- $\boldsymbol{\mu} = (\mu_1, \mu_2, \ldots, \mu_n)^T$ be the vector of expected returns
- $\boldsymbol{\Sigma}$ be the $n \times n$ covariance matrix of returns

### 3.2 Portfolio Return and Risk

**Expected Portfolio Return:**
$$\mu_p = \mathbf{w}^T \boldsymbol{\mu} = \sum_{i=1}^{n} w_i \mu_i$$

**Portfolio Variance (Risk):**
$$\sigma_p^2 = \mathbf{w}^T \boldsymbol{\Sigma} \mathbf{w} = \sum_{i=1}^{n} \sum_{j=1}^{n} w_i w_j \sigma_{ij}$$

where $\sigma_{ij}$ is the covariance between assets $i$ and $j$, and $\sigma_{ii} = \sigma_i^2$ is the variance of asset $i$.

### 3.3 The Optimization Problem

The Markowitz optimization problem can be formulated in several equivalent ways:

**Formulation 1: Minimize risk for a target return**
$$\min_{\mathbf{w}} \quad \mathbf{w}^T \boldsymbol{\Sigma} \mathbf{w}$$
$$\text{subject to:} \quad \mathbf{w}^T \boldsymbol{\mu} = \mu_{target}$$
$$\mathbf{w}^T \mathbf{1} = 1$$

**Formulation 2: Maximize return for a target risk**
$$\max_{\mathbf{w}} \quad \mathbf{w}^T \boldsymbol{\mu}$$
$$\text{subject to:} \quad \mathbf{w}^T \boldsymbol{\Sigma} \mathbf{w} \leq \sigma_{target}^2$$
$$\mathbf{w}^T \mathbf{1} = 1$$

**Formulation 3: Mean-Variance Utility**
$$\max_{\mathbf{w}} \quad \mathbf{w}^T \boldsymbol{\mu} - \frac{\lambda}{2} \mathbf{w}^T \boldsymbol{\Sigma} \mathbf{w}$$
$$\text{subject to:} \quad \mathbf{w}^T \mathbf{1} = 1$$

where $\lambda > 0$ is the risk aversion parameter.

### 3.4 The Efficient Frontier

The **efficient frontier** is the set of optimal portfolios that offer the highest expected return for a defined level of risk (or the lowest risk for a given expected return). Portfolios below the efficient frontier are sub-optimal because they don't provide enough return for their level of risk.

## 4. Model Assumptions <a id='4-model-assumptions'></a>

The Markowitz framework relies on several key assumptions:

### 4.1 Assumptions About Investors

1. **Risk Aversion:** Investors are risk-averse; they prefer less risk to more risk for the same level of expected return.

2. **Mean-Variance Preferences:** Investors make decisions based solely on the mean (expected return) and variance (risk) of portfolio returns. This implies either:
   - Returns are normally distributed, OR
   - Investors have quadratic utility functions

3. **Single-Period Investment Horizon:** Investors optimize for a single, common investment period.

4. **Rational Behavior:** Investors act rationally and seek to maximize their expected utility.

### 4.2 Assumptions About Markets

5. **Frictionless Markets:** No transaction costs, taxes, or restrictions on short selling.

6. **Divisibility:** Assets are infinitely divisible (investors can buy any fraction of an asset).

7. **Price Takers:** Investors are price takers; their trades do not affect market prices.

8. **Perfect Information:** All investors have access to the same information and have the same estimates of expected returns, variances, and covariances.

### 4.3 Assumptions About Returns

9. **Known Parameters:** The expected returns, variances, and covariances are known (or can be accurately estimated).

10. **Stationarity:** The statistical properties of returns remain constant over time.

## 5. Strengths of the Markowitz Framework <a id='5-strengths'></a>

### 5.1 Theoretical Contributions

1. **Formalization of Diversification:** Markowitz mathematically demonstrated why "not putting all eggs in one basket" reduces risk. The key insight is that portfolio risk depends on correlations between assets, not just individual asset risks.

2. **Risk-Return Trade-off:** The framework provides a clear, quantitative relationship between risk and return, enabling objective portfolio comparison.

3. **Foundation for Modern Finance:** MPT laid the groundwork for CAPM, APT, and virtually all modern portfolio management techniques.

### 5.2 Practical Advantages

4. **Systematic Approach:** Provides a disciplined, mathematical approach to portfolio construction rather than relying on intuition.

5. **Quantitative Framework:** Enables the use of optimization algorithms and computer-based portfolio management.

6. **Flexibility:** Can incorporate constraints such as:
   - No short selling ($w_i \geq 0$)
   - Position limits ($w_i \leq w_{max}$)
   - Sector constraints
   - ESG requirements

7. **Intuitive Output:** The efficient frontier is a visually intuitive tool for understanding risk-return trade-offs.

8. **Benchmark for Comparison:** Even if not used directly, the efficient frontier serves as a benchmark against which other portfolios can be evaluated.

---

## 6. Weaknesses and Limitations <a id='6-weaknesses'></a>

### 6.1 Estimation Error

1. **Parameter Sensitivity:** The optimization is extremely sensitive to input estimates. Small changes in expected returns can lead to dramatically different optimal portfolios (Michaud, 1989).

2. **Estimation Risk:** Expected returns are notoriously difficult to estimate accurately. Historical returns are poor predictors of future returns.

3. **Error Maximization:** Optimization tends to overweight assets with high estimated returns (which may be estimation errors) and underweight assets with low estimated returns.

### 6.2 Statistical Limitations

4. **Non-Normal Returns:** Asset returns often exhibit fat tails, skewness, and other non-normal characteristics. Variance may not adequately capture risk.

5. **Time-Varying Parameters:** Correlations and volatilities change over time, especially during market stress (correlation breakdown).

6. **Single-Period Model:** Real investment is multi-period, requiring consideration of rebalancing, changing circumstances, and intermediate consumption.

### 6.3 Practical Limitations

7. **Extreme Weights:** Without constraints, optimization often produces extreme long and short positions that are impractical.

8. **Transaction Costs:** The model ignores trading costs, which can significantly erode returns, especially for frequent rebalancing.

9. **Illiquidity:** Some assets cannot be traded in arbitrary quantities without market impact.

10. **Concentration Risk:** Optimized portfolios may be highly concentrated in a few assets, leading to idiosyncratic risk.

### 6.4 Behavioral Limitations

11. **Utility Assumptions:** Quadratic utility implies increasing absolute risk aversion (investors become more risk-averse as wealth increases), which contradicts observed behavior.

12. **Loss Aversion:** Behavioral finance shows that investors are more sensitive to losses than gains (prospect theory), which is not captured by symmetric variance.

### 7.1 Generate Synthetic Asset Returns

```python
# Define parameters for 5 synthetic assets
n_assets = 5
n_observations = 252 * 5  # 5 years of daily data

# Expected annual returns (in decimal form)
expected_returns = np.array([0.12, 0.10, 0.15, 0.08, 0.18])  # 12%, 10%, 15%, 8%, 18%

# Annual volatilities
volatilities = np.array([0.20, 0.15, 0.25, 0.10, 0.30])  # 20%, 15%, 25%, 10%, 30%

# Correlation matrix
correlation_matrix = np.array([
    [1.00, 0.30, 0.50, 0.10, 0.40],
    [0.30, 1.00, 0.25, 0.20, 0.35],
    [0.50, 0.25, 1.00, 0.15, 0.60],
    [0.10, 0.20, 0.15, 1.00, 0.05],
    [0.40, 0.35, 0.60, 0.05, 1.00]
])

# Convert to covariance matrix
cov_matrix = np.outer(volatilities, volatilities) * correlation_matrix

# Asset names
asset_names = ['Tech Stock', 'Utility Stock', 'Growth Stock', 'Bond ETF', 'Small Cap']

print("Synthetic Asset Parameters:")
print("="*50)
params_df = pd.DataFrame({
    'Asset': asset_names,
    'Expected Return': [f"{r*100:.1f}%" for r in expected_returns],
    'Volatility': [f"{v*100:.1f}%" for v in volatilities]
})
print(params_df.to_string(index=False))

print("\nCorrelation Matrix:")
corr_df = pd.DataFrame(correlation_matrix, index=asset_names, columns=asset_names)
print(corr_df.round(2))
```

```python
# Generate synthetic daily returns using multivariate normal distribution
daily_mean = expected_returns / 252  # Convert to daily
daily_cov = cov_matrix / 252  # Convert to daily

# Generate returns
synthetic_returns = np.random.multivariate_normal(daily_mean, daily_cov, n_observations)
returns_df = pd.DataFrame(synthetic_returns, columns=asset_names)

# Create date index
dates = pd.date_range(start='2019-01-01', periods=n_observations, freq='B')
returns_df.index = dates

print("Generated Daily Returns (first 10 rows):")
print(returns_df.head(10).round(4))

print("\nSample Statistics (Annualized):")
stats_df = pd.DataFrame({
    'Mean Return': returns_df.mean() * 252,
    'Volatility': returns_df.std() * np.sqrt(252),
    'Sharpe Ratio': (returns_df.mean() * 252) / (returns_df.std() * np.sqrt(252))
})
print(stats_df.round(4))
```

```python
# Calculate cumulative returns (price paths)
cumulative_returns = (1 + returns_df).cumprod()

# Plot cumulative returns
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Plot 1: Cumulative Returns
ax1 = axes[0, 0]
for col in cumulative_returns.columns:
    ax1.plot(cumulative_returns.index, cumulative_returns[col], label=col, linewidth=1.5)
ax1.set_title('Cumulative Returns (Synthetic Data)', fontsize=12, fontweight='bold')
ax1.set_xlabel('Date')
ax1.set_ylabel('Cumulative Return (Starting at $1)')
ax1.legend(loc='upper left')
ax1.grid(True, alpha=0.3)

# Plot 2: Return Distributions
ax2 = axes[0, 1]
returns_df.plot(kind='hist', bins=50, alpha=0.5, ax=ax2)
ax2.set_title('Return Distributions', fontsize=12, fontweight='bold')
ax2.set_xlabel('Daily Return')
ax2.set_ylabel('Frequency')
ax2.legend(loc='upper right')

# Plot 3: Correlation Heatmap
ax3 = axes[1, 0]
sample_corr = returns_df.corr()
im = ax3.imshow(sample_corr, cmap='RdYlBu_r', aspect='auto', vmin=-1, vmax=1)
ax3.set_xticks(range(len(asset_names)))
ax3.set_yticks(range(len(asset_names)))
ax3.set_xticklabels(asset_names, rotation=45, ha='right')
ax3.set_yticklabels(asset_names)
ax3.set_title('Sample Correlation Matrix', fontsize=12, fontweight='bold')
for i in range(len(asset_names)):
    for j in range(len(asset_names)):
        ax3.text(j, i, f'{sample_corr.iloc[i, j]:.2f}', ha='center', va='center', fontsize=9)
plt.colorbar(im, ax=ax3)

# Plot 4: Risk-Return Scatter of Individual Assets
ax4 = axes[1, 1]
ann_returns = returns_df.mean() * 252
ann_vol = returns_df.std() * np.sqrt(252)
colors = plt.cm.Set2(np.linspace(0, 1, len(asset_names)))
for i, name in enumerate(asset_names):
    ax4.scatter(ann_vol[name], ann_returns[name], s=200, c=[colors[i]], 
                label=name, edgecolors='black', linewidth=1.5)
ax4.set_title('Risk-Return Profile of Individual Assets', fontsize=12, fontweight='bold')
ax4.set_xlabel('Annualized Volatility')
ax4.set_ylabel('Annualized Return')
ax4.legend(loc='upper left')
ax4.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

### 7.3 Implement Markowitz Optimization

```python
class MarkowitzOptimizer:
    """
    Implementation of Markowitz Mean-Variance Portfolio Optimization.
    """
    
    def __init__(self, returns, risk_free_rate=0.02):
        """
        Initialize the optimizer.
        
        Parameters:
        -----------
        returns : pd.DataFrame
            DataFrame of asset returns (daily or periodic)
        risk_free_rate : float
            Annual risk-free rate (default: 2%)
        """
        self.returns = returns
        self.n_assets = returns.shape[1]
        self.asset_names = returns.columns.tolist()
        self.risk_free_rate = risk_free_rate
        
        # Calculate annualized statistics
        self.mean_returns = returns.mean() * 252
        self.cov_matrix = returns.cov() * 252
        
    def portfolio_return(self, weights):
        """Calculate expected portfolio return."""
        return np.dot(weights, self.mean_returns)
    
    def portfolio_volatility(self, weights):
        """Calculate portfolio volatility (standard deviation)."""
        return np.sqrt(np.dot(weights.T, np.dot(self.cov_matrix, weights)))
    
    def portfolio_sharpe(self, weights):
        """Calculate portfolio Sharpe ratio."""
        ret = self.portfolio_return(weights)
        vol = self.portfolio_volatility(weights)
        return (ret - self.risk_free_rate) / vol
    
    def minimize_volatility(self, target_return, allow_short=False):
        """
        Find the minimum volatility portfolio for a given target return.
        """
        n = self.n_assets
        
        # Constraints
        constraints = [
            {'type': 'eq', 'fun': lambda w: np.sum(w) - 1},
            {'type': 'eq', 'fun': lambda w: self.portfolio_return(w) - target_return}
        ]
        
        # Bounds
        if allow_short:
            bounds = tuple((-1, 1) for _ in range(n))
        else:
            bounds = tuple((0, 1) for _ in range(n))
        
        # Initial guess (equal weights)
        w0 = np.ones(n) / n
        
        # Optimize
        result = minimize(
            self.portfolio_volatility,
            w0,
            method='SLSQP',
            bounds=bounds,
            constraints=constraints
        )
        
        return result.x if result.success else None
    
    def maximize_sharpe(self, allow_short=False):
        """
        Find the maximum Sharpe ratio portfolio (tangency portfolio).
        """
        n = self.n_assets
        
        # Constraints
        constraints = [{'type': 'eq', 'fun': lambda w: np.sum(w) - 1}]
        
        # Bounds
        if allow_short:
            bounds = tuple((-1, 1) for _ in range(n))
        else:
            bounds = tuple((0, 1) for _ in range(n))
        
        # Initial guess
        w0 = np.ones(n) / n
        
        # Optimize (minimize negative Sharpe ratio)
        result = minimize(
            lambda w: -self.portfolio_sharpe(w),
            w0,
            method='SLSQP',
            bounds=bounds,
            constraints=constraints
        )
        
        return result.x if result.success else None
    
    def minimum_variance_portfolio(self, allow_short=False):
        """
        Find the global minimum variance portfolio.
        """
        n = self.n_assets
        
        # Constraints
        constraints = [{'type': 'eq', 'fun': lambda w: np.sum(w) - 1}]
        
        # Bounds
        if allow_short:
            bounds = tuple((-1, 1) for _ in range(n))
        else:
            bounds = tuple((0, 1) for _ in range(n))
        
        # Initial guess
        w0 = np.ones(n) / n
        
        # Optimize
        result = minimize(
            self.portfolio_volatility,
            w0,
            method='SLSQP',
            bounds=bounds,
            constraints=constraints
        )
        
        return result.x if result.success else None
    
    def efficient_frontier(self, n_points=100, allow_short=False):
        """
        Calculate the efficient frontier.
        """
        # Find minimum and maximum returns on the frontier
        min_ret = self.mean_returns.min()
        max_ret = self.mean_returns.max()
        
        # Get minimum variance portfolio return
        mvp_weights = self.minimum_variance_portfolio(allow_short)
        mvp_return = self.portfolio_return(mvp_weights)
        
        # Efficient frontier is above MVP
        target_returns = np.linspace(mvp_return, max_ret, n_points)
        
        frontier_vol = []
        frontier_ret = []
        frontier_weights = []
        
        for target in target_returns:
            weights = self.minimize_volatility(target, allow_short)
            if weights is not None:
                vol = self.portfolio_volatility(weights)
                ret = self.portfolio_return(weights)
                frontier_vol.append(vol)
                frontier_ret.append(ret)
                frontier_weights.append(weights)
        
        return np.array(frontier_ret), np.array(frontier_vol), np.array(frontier_weights)

print("MarkowitzOptimizer class defined successfully!")
```

### 7.4 Calculate and Plot the Efficient Frontier

```python
# Create optimizer
optimizer = MarkowitzOptimizer(returns_df, risk_free_rate=0.02)

# Calculate efficient frontier
ef_returns, ef_vols, ef_weights = optimizer.efficient_frontier(n_points=100, allow_short=False)

# Find special portfolios
mvp_weights = optimizer.minimum_variance_portfolio(allow_short=False)
msr_weights = optimizer.maximize_sharpe(allow_short=False)

mvp_return = optimizer.portfolio_return(mvp_weights)
mvp_vol = optimizer.portfolio_volatility(mvp_weights)
mvp_sharpe = optimizer.portfolio_sharpe(mvp_weights)

msr_return = optimizer.portfolio_return(msr_weights)
msr_vol = optimizer.portfolio_volatility(msr_weights)
msr_sharpe = optimizer.portfolio_sharpe(msr_weights)

# Equal weight portfolio
ew_weights = np.ones(optimizer.n_assets) / optimizer.n_assets
ew_return = optimizer.portfolio_return(ew_weights)
ew_vol = optimizer.portfolio_volatility(ew_weights)
ew_sharpe = optimizer.portfolio_sharpe(ew_weights)

print("Portfolio Comparison:")
print("="*70)
comparison_df = pd.DataFrame({
    'Portfolio': ['Minimum Variance', 'Maximum Sharpe', 'Equal Weight'],
    'Expected Return': [f"{mvp_return*100:.2f}%", f"{msr_return*100:.2f}%", f"{ew_return*100:.2f}%"],
    'Volatility': [f"{mvp_vol*100:.2f}%", f"{msr_vol*100:.2f}%", f"{ew_vol*100:.2f}%"],
    'Sharpe Ratio': [f"{mvp_sharpe:.3f}", f"{msr_sharpe:.3f}", f"{ew_sharpe:.3f}"]
})
print(comparison_df.to_string(index=False))
```

### 7.5 Sensitivity Analysis: The Impact of Estimation Error

```python
# Demonstrate sensitivity to expected return estimates
np.random.seed(123)

# Original expected returns
original_returns = optimizer.mean_returns.copy()

# Perturb expected returns slightly
n_simulations = 50
perturbation_std = 0.02  # 2% perturbation

perturbed_weights = []
perturbed_frontiers = []

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

ax1 = axes[0]
ax2 = axes[1]

# Original efficient frontier
ax1.plot(ef_vols, ef_returns, 'b-', linewidth=3, label='Original Frontier', zorder=5)

for sim in range(n_simulations):
    # Perturb returns
    noise = np.random.normal(0, perturbation_std, len(original_returns))
    perturbed_mean_returns = original_returns + noise
    
    # Create new optimizer with perturbed returns
    perturbed_optimizer = MarkowitzOptimizer(returns_df, risk_free_rate=0.02)
    perturbed_optimizer.mean_returns = pd.Series(perturbed_mean_returns, index=asset_names)
    
    # Calculate perturbed frontier
    try:
        p_ret, p_vol, p_weights = perturbed_optimizer.efficient_frontier(n_points=50, allow_short=False)
        ax1.plot(p_vol, p_ret, 'gray', alpha=0.2, linewidth=1)
        
        # Get max Sharpe portfolio weights
        msr_w = perturbed_optimizer.maximize_sharpe(allow_short=False)
        if msr_w is not None:
            perturbed_weights.append(msr_w)
    except:
        continue

ax1.scatter(msr_vol, msr_return, marker='*', s=400, c='green', 
            edgecolors='black', linewidth=2, zorder=6, label='Original Max Sharpe')
ax1.set_xlabel('Annualized Volatility', fontsize=11)
ax1.set_ylabel('Annualized Return', fontsize=11)
ax1.set_title(f'Efficient Frontier Sensitivity\n(+/-{perturbation_std*100:.0f}% Return Estimation Error)', 
              fontsize=12, fontweight='bold')
ax1.legend(loc='upper left')
ax1.grid(True, alpha=0.3)

# Plot weight variability
perturbed_weights = np.array(perturbed_weights)
weight_df = pd.DataFrame(perturbed_weights, columns=asset_names)

bp = ax2.boxplot([weight_df[col].values for col in asset_names], labels=asset_names)
ax2.scatter(range(1, len(asset_names)+1), msr_weights, marker='*', s=200, 
            c='green', zorder=5, label='Original Weights')
ax2.set_ylabel('Portfolio Weight', fontsize=11)
ax2.set_title('Max Sharpe Portfolio Weight Variability\nDue to Estimation Error', 
              fontsize=12, fontweight='bold')
ax2.tick_params(axis='x', rotation=45)
ax2.legend(loc='upper right')
ax2.grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.show()

print("\nWeight Statistics from Sensitivity Analysis:")
print("="*60)
sensitivity_stats = pd.DataFrame({
    'Asset': asset_names,
    'Original Weight': [f"{w*100:.1f}%" for w in msr_weights],
    'Mean Weight': [f"{weight_df[col].mean()*100:.1f}%" for col in asset_names],
    'Std Weight': [f"{weight_df[col].std()*100:.1f}%" for col in asset_names],
    'Min Weight': [f"{weight_df[col].min()*100:.1f}%" for col in asset_names],
    'Max Weight': [f"{weight_df[col].max()*100:.1f}%" for col in asset_names]
})
print(sensitivity_stats.to_string(index=False))
```

Now let's apply the Markowitz framework to real market data. We'll use historical stock prices from well-known companies.

```python
# Define tickers - a diversified mix of stocks
tickers = ['AAPL', 'MSFT', 'JNJ', 'JPM', 'XOM', 'PG']
ticker_names = {
    'AAPL': 'Apple Inc.',
    'MSFT': 'Microsoft Corp.',
    'JNJ': 'Johnson & Johnson',
    'JPM': 'JPMorgan Chase',
    'XOM': 'Exxon Mobil',
    'PG': 'Procter & Gamble'
}

if USE_REAL_DATA:
    # Fetch real data from Yahoo Finance
    print("Fetching data from Yahoo Finance...")
    start_date = '2019-01-01'
    end_date = '2024-01-01'
    
    data = yf.download(tickers, start=start_date, end=end_date, auto_adjust=True)['Close']
    yf_returns = data.pct_change().dropna()
    print(f"Fetched {len(yf_returns)} days of data for {len(tickers)} stocks.")
else:
    # Simulate realistic stock data based on typical market characteristics
    print("Generating simulated Yahoo Finance-style data...")
    
    # Realistic parameters based on historical market data (2019-2023 period)
    sim_returns = np.array([0.25, 0.28, 0.08, 0.12, 0.05, 0.10])
    sim_volatilities = np.array([0.30, 0.28, 0.18, 0.25, 0.28, 0.15])
    
    # Correlation matrix (realistic sector correlations)
    sim_correlation = np.array([
        [1.00, 0.75, 0.30, 0.45, 0.20, 0.35],
        [0.75, 1.00, 0.35, 0.50, 0.25, 0.40],
        [0.30, 0.35, 1.00, 0.40, 0.30, 0.55],
        [0.45, 0.50, 0.40, 1.00, 0.50, 0.45],
        [0.20, 0.25, 0.30, 0.50, 1.00, 0.25],
        [0.35, 0.40, 0.55, 0.45, 0.25, 1.00]
    ])
    
    # Generate covariance matrix
    sim_cov = np.outer(sim_volatilities, sim_volatilities) * sim_correlation
    
    # Generate 5 years of daily data
    n_days = 252 * 5
    daily_mean = sim_returns / 252
    daily_cov = sim_cov / 252
    
    np.random.seed(2024)
    simulated_returns = np.random.multivariate_normal(daily_mean, daily_cov, n_days)
    
    # Create DataFrame
    dates = pd.date_range(start='2019-01-01', periods=n_days, freq='B')
    yf_returns = pd.DataFrame(simulated_returns, index=dates, columns=tickers)
    
    print(f"Generated {len(yf_returns)} days of simulated data for {len(tickers)} stocks.")

print("\nFirst 5 rows of daily returns:")
print(yf_returns.head().round(4))
```

### 8.2 Backtesting the Optimized Portfolios

```python
# Simple backtest: compare portfolio performance over time
# Calculate portfolio returns
mvp_portfolio_returns = (yf_returns * yf_mvp_weights).sum(axis=1)
msr_portfolio_returns = (yf_returns * yf_msr_weights).sum(axis=1)
ew_portfolio_returns = (yf_returns * yf_ew_weights).sum(axis=1)

# Create portfolio DataFrame
portfolio_returns = pd.DataFrame({
    'Minimum Variance': mvp_portfolio_returns,
    'Maximum Sharpe': msr_portfolio_returns,
    'Equal Weight': ew_portfolio_returns
})

# Calculate cumulative returns
portfolio_cumulative = (1 + portfolio_returns).cumprod()

# Plot backtest results
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Plot 1: Cumulative Returns
ax1 = axes[0, 0]
for col in portfolio_cumulative.columns:
    ax1.plot(portfolio_cumulative.index, portfolio_cumulative[col], label=col, linewidth=2)
ax1.set_title('Cumulative Portfolio Returns', fontsize=12, fontweight='bold')
ax1.set_xlabel('Date')
ax1.set_ylabel('Cumulative Return ($1 invested)')
ax1.legend(loc='upper left')
ax1.grid(True, alpha=0.3)

# Plot 2: Rolling Volatility
ax2 = axes[0, 1]
rolling_window = 63
rolling_vol = portfolio_returns.rolling(window=rolling_window).std() * np.sqrt(252)
for col in rolling_vol.columns:
    ax2.plot(rolling_vol.index, rolling_vol[col], label=col, linewidth=1.5)
ax2.set_title(f'Rolling {rolling_window}-Day Volatility', fontsize=12, fontweight='bold')
ax2.set_xlabel('Date')
ax2.set_ylabel('Annualized Volatility')
ax2.legend(loc='upper right')
ax2.grid(True, alpha=0.3)

# Plot 3: Drawdown
ax3 = axes[1, 0]
for col in portfolio_cumulative.columns:
    running_max = portfolio_cumulative[col].expanding().max()
    drawdown = (portfolio_cumulative[col] - running_max) / running_max
    ax3.fill_between(drawdown.index, drawdown, alpha=0.3, label=col)
    ax3.plot(drawdown.index, drawdown, linewidth=1.5)
ax3.set_title('Portfolio Drawdown', fontsize=12, fontweight='bold')
ax3.set_xlabel('Date')
ax3.set_ylabel('Drawdown')
ax3.legend(loc='lower left')
ax3.grid(True, alpha=0.3)

# Plot 4: Performance Metrics Bar Chart
ax4 = axes[1, 1]

# Calculate metrics
metrics = pd.DataFrame()
for col in portfolio_returns.columns:
    total_ret = portfolio_cumulative[col].iloc[-1] - 1
    ann_ret = (1 + total_ret) ** (252 / len(portfolio_returns)) - 1
    ann_vol = portfolio_returns[col].std() * np.sqrt(252)
    sharpe = (ann_ret - 0.02) / ann_vol
    max_dd = ((portfolio_cumulative[col] - portfolio_cumulative[col].expanding().max()) / 
              portfolio_cumulative[col].expanding().max()).min()
    
    metrics[col] = [ann_ret, ann_vol, sharpe, max_dd]

metrics.index = ['Annual Return', 'Volatility', 'Sharpe Ratio', 'Max Drawdown']

x = np.arange(len(metrics.index))
width = 0.25
multiplier = 0

colors = ['#2E86AB', '#A23B72', '#F18F01']
for i, (col, color) in enumerate(zip(portfolio_returns.columns, colors)):
    offset = width * multiplier
    bars = ax4.bar(x + offset, metrics[col], width, label=col, color=color, alpha=0.8)
    multiplier += 1

ax4.set_ylabel('Value', fontsize=11)
ax4.set_title('Performance Metrics Comparison', fontsize=12, fontweight='bold')
ax4.set_xticks(x + width)
ax4.set_xticklabels(metrics.index, fontsize=10)
ax4.legend(loc='upper right')
ax4.axhline(y=0, color='black', linewidth=0.5)
ax4.grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.show()

print("\nBacktest Performance Summary:")
print("="*60)
print(metrics.round(4).to_string())
```

## 9. Extensions and Modern Developments <a id='9-extensions'></a>

The Markowitz framework has inspired numerous extensions to address its limitations:

### 9.1 Robust Optimization
Addresses estimation error by optimizing for the worst-case scenario within uncertainty sets for parameters (Goldfarb & Iyengar, 2003).

### 9.2 Black-Litterman Model
Combines market equilibrium returns with investor views to produce more stable estimates (Black & Litterman, 1992).

### 9.3 Risk Parity
Equalizes risk contributions from each asset rather than optimizing return-to-risk ratio (Qian, 2005).

### 9.4 Bayesian Methods
Uses Bayesian inference to better account for parameter uncertainty (Pastor, 2000).

### 9.5 Machine Learning Approaches
Uses ML techniques for better return prediction and covariance estimation (Lopez de Prado, 2018).

### 9.6 Factor Models
Reduces dimensionality by modeling covariance through common factors (Ross, 1976; Fama & French, 1993).

### 9.7 Hierarchical Risk Parity
Uses hierarchical clustering to build diversified portfolios that are more robust to estimation error (Lopez de Prado, 2016).

## 10. Conclusion <a id='10-conclusion'></a>

The Markowitz Portfolio Optimization framework remains a cornerstone of modern finance despite being over 70 years old. Its key contributions include:

1. **Formalizing diversification** - Demonstrating mathematically why diversification reduces risk
2. **Efficient frontier concept** - Providing a clear framework for understanding risk-return trade-offs
3. **Foundation for modern finance** - Enabling the development of CAPM, APT, and countless other advances

However, practitioners should be aware of its limitations:

1. **Sensitivity to inputs** - Small changes in estimates can lead to dramatically different portfolios
2. **Estimation challenges** - Expected returns are notoriously difficult to estimate accurately
3. **Model assumptions** - Real markets violate many of the theory's assumptions

Modern portfolio management typically combines Markowitz optimization with additional techniques such as regularization and constraints, robust optimization methods, shrinkage estimators for covariance matrices, and resampling techniques (Michaud, 1998).

Understanding both the power and limitations of the Markowitz framework is essential for any serious practitioner in quantitative finance.
